```python
import pandas as pd
a = pd.read_csv("admission_predict.csv")
print("******************** D A T A  L O A D I N G *********************")
print(a)
```

```python
print("********************D A T A  S P I L I T T I N G **************************")
x=a.iloc[:,:-1].values
print(x)
y=a["Research"]
print(y)
```

```python
print("***********************D A T A  S P I L I T T I N G **************************")
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,shuffle=True)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
print("*********************D A T A   C L A S S I F I C A T I O N **********************")
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
s=LogisticRegression()
s.fit(x_train,y_train)
print("************* L O G I S T I C  R E G R E S S I O N ****************")
prediction=s.predict(x_test)
print("accuracy",accuracy_score(y_test,prediction))
```

```python
print("################# C O N F U S I O N  M A T R I X ################")
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction)
```

```python
print("$$$$$$$$$$$$$$$$$$$$$ C L A S S I F I C A T I O N  R E P O R T $$$$$$$$$$$$$$$$$$$$$$")
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```

```python
print("************* D E C I S I O N  T R E E ****************")
from sklearn.tree import DecisionTreeClassifier
s=DecisionTreeClassifier()
s.fit(x_train,y_train)
prediction=s.predict(x_test)
print("accuracy",accuracy_score(y_test,prediction))
```

```python
print("################# C O N F U S I O N  M A T R I X ################")
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction)
```

```python
print("$$$$$$$$$$$$$$$$$$$$$ C L A S S I F I C A T I O N  R E P O R T $$$$$$$$$$$$$$$$$$$$$$")
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```
